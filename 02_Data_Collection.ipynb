```python
# Imports
import pandas as pd

import requests
from bs4 import BeautifulSoup
import wikipedia
```

## Import Musicals List

```python
musical_names = pd.read_csv('./data/musical_names.csv')
musical_names.shape
```

```python
musical_names.head()
```

# Get Musical Summaries

```python
# Credits (auto_suggest): @nielstron, 10.10.2018
# https://github.com/goldsmith/Wikipedia/issues/192

# Get Wikipedia pages for all 196 musicals
summaries = []
i = 0
for musical in musical_names['musical']:
    name_to_search = wikipedia.search(musical.strip().lower() + ' musical')[0]
    print(name_to_search)
    summary = wikipedia.page(name_to_search, auto_suggest=False).content
    summaries.append(summary)
    i += 1
    print(f'Summary #{i} added. Continuing to summary #{i+1}...')

len(summaries)
```

```python
# Add summaries to dataframe
musical_names['description'] = summaries
musical_names.head()
```

# Export Data

```python
musical_names.to_csv('./data/musical_summaries.csv', index=False)
```
